# D2 â€” Quivers and counting paths / Quivers y conteo de caminos

đź‡¬đź‡§ A **quiver** is just a directed graph that is allowed to have several arrows between the same two nodes.
Nothing scary. What makes it powerful is that we can *count paths* in it, and that counting follows clean
algebraic rules â€” the **path algebra** `kQ`. This notebook builds the intuition with the smallest possible
example, by hand.

đź‡Şđź‡¸ Un **quiver** es simplemente un grafo dirigido al que se le permite tener varias flechas entre los mismos
dos nodos. Nada que temer. Lo que lo hace poderoso es que podemos *contar caminos* en Ă©l, y ese conteo sigue
reglas algebraicas limpias â€” el **Ăˇlgebra de caminos** `kQ`. Este cuaderno construye la intuiciĂłn con el
ejemplo mĂˇs pequeĂ±o posible, a mano.

## The diamond / El diamante

đź‡¬đź‡§ Four nodes. Two ways to get from node 0 to node 3 in two steps: `0â†’1â†’3` and `0â†’2â†’3`. Hold that thought â€”
*two paths* â€” we'll recover it algebraically.

đź‡Şđź‡¸ Cuatro nodos. Dos maneras de ir del nodo 0 al nodo 3 en dos pasos: `0â†’1â†’3` y `0â†’2â†’3`. Recuerda eso â€”
*dos caminos* â€” lo recuperaremos algebraicamente.

```python
import numpy as np, networkx as nx, matplotlib.pyplot as plt
from aiq.quiver import Quiver

# arrows: (name, from, to) / flechas: (nombre, desde, hacia)
Q = Quiver([0,1,2,3], [('a',0,1), ('b',0,2), ('c',1,3), ('d',2,3)])
G = nx.DiGraph(); G.add_edges_from([(0,1),(0,2),(1,3),(2,3)])
pos = {0:(0,0), 1:(1,1), 2:(1,-1), 3:(2,0)}
plt.figure(figsize=(5,3))
nx.draw(G, pos, with_labels=True, node_color='#93c5fd', node_size=800, arrowsize=16)
nx.draw_networkx_edge_labels(G, pos, {(0,1):'a',(0,2):'b',(1,3):'c',(2,3):'d'})
plt.title('the diamond quiver'); plt.show()
```

## Counting paths two ways / Contar caminos de dos formas

đź‡¬đź‡§ **Way 1 â€” by hand / with the algebra.** The path algebra knows every directed path. Ask it for all paths
of length 2 from 0 to 3.

đź‡Şđź‡¸ **Forma 1 â€” a mano / con el Ăˇlgebra.** El Ăˇlgebra de caminos conoce todo camino dirigido. PĂ­dele todos
los caminos de longitud 2 de 0 a 3.

```python
from aiq.path_algebra import PathAlgebra
alg = PathAlgebra(Q)
paths = alg.paths_from_to(0, 3, 2)
print('paths of length 2 from 0 to 3:')
for p in paths: print('   ', p)
print(f'\ncount = {len(paths)}')
```

đź‡¬đź‡§ **Way 2 â€” the adjacency matrix.** Here is the magic that ties graphs to algebra: if `A` is the adjacency
matrix (`A[i,j]` = number of arrows `iâ†’j`), then `(A^k)[i,j]` is **exactly the number of length-`k` paths from
`i` to `j`**. No enumeration needed â€” just multiply matrices.

đź‡Şđź‡¸ **Forma 2 â€” la matriz de adyacencia.** AquĂ­ estĂˇ la magia que une grafos con Ăˇlgebra: si `A` es la matriz
de adyacencia (`A[i,j]` = nĂşmero de flechas `iâ†’j`), entonces `(A^k)[i,j]` es **exactamente el nĂşmero de
caminos de longitud `k` de `i` a `j`**. Sin enumerar â€” solo multiplica matrices.

```python
A = Q.adjacency_matrix()
print('A =\n', A.astype(int))
A2 = A @ A
print('\nA^2 =\n', A2.astype(int))
print(f'\n(A^2)[0,3] = {int(A2[0,3])}   <-- same as the count above!')
assert int(A2[0,3]) == len(paths)
print('Algebra and matrix agree. / El Ăˇlgebra y la matriz coinciden.')
```

## The notation we will use / La notaciĂłn que usaremos

đź‡¬đź‡§ In the thesis this number is written
$$n_g(i,j) = \dim_k\big(e_i \cdot kQ_g \cdot e_j\big) = (A^g)_{ij},$$
â€śthe number of length-`g` walks from `i` to `j`â€ť. You do **not** need the formal definition â€” just remember
the plain-English version: **`n_g(i,j)` counts the ways to get from `i` to `j` in exactly `g` steps.**

đź‡Şđź‡¸ En la tesis este nĂşmero se escribe
$$n_g(i,j) = \dim_k\big(e_i \cdot kQ_g \cdot e_j\big) = (A^g)_{ij},$$
â€śel nĂşmero de recorridos de longitud `g` de `i` a `j`â€ť. **No** necesitas la definiciĂłn formal â€” solo recuerda
la versiĂłn en lenguaje llano: **`n_g(i,j)` cuenta las maneras de ir de `i` a `j` en exactamente `g` pasos.**

## Why this matters next / Por quĂ© esto importa despuĂ©s

đź‡¬đź‡§ We now have an exact handle on *how many paths* connect any two nodes at any distance. In **D3** we'll see
that **each path is a message**, so `n_g(i,j)` is literally the number of messages that pile up â€” the thing
that gets squashed.

đź‡Şđź‡¸ Ahora tenemos un control exacto de *cuĂˇntos caminos* conectan dos nodos a cualquier distancia. En **D3**
veremos que **cada camino es un mensaje**, asĂ­ que `n_g(i,j)` es literalmente el nĂşmero de mensajes que se
acumulan â€” lo que se aprieta.
