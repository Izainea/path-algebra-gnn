# P2 — Quivers and counting paths

A **quiver** is a directed graph (multiple arrows allowed). Its power: we can *count paths* with clean
algebra. The key fact — **`(A^g)[i,j]` = number of length-`g` paths from `i` to `j`** — ties graphs to the
path algebra `kQ`.

**Figures:** (1) the diamond quiver, (2) `A` vs `A²` as matrices, (3) paths-per-length bar chart.

```python
import numpy as np
from oversquash import viz
from aiq.quiver import Quiver
from aiq.path_algebra import PathAlgebra
Q = Quiver([0,1,2,3], [('a',0,1),('b',0,2),('c',1,3),('d',2,3)])
A = Q.adjacency_matrix()
```

## Figure 1 — the diamond quiver

Two ways from `0` to `3` in two steps: `0→1→3` and `0→2→3`. Hold that — *two paths*.

```python
viz.draw_graph([(0,1),(0,2),(1,3),(2,3)], 4, src=0, dst=3,
               pos={0:(0,0),1:(1,1),2:(1,-1),3:(2,0)}, title='diamond quiver')
paths = PathAlgebra(Q).paths_from_to(0, 3, 2)
print('paths of length 2 from 0 to 3:', [str(p) for p in paths], '-> count =', len(paths))
```

## Figure 2 — the adjacency matrix and its square

`A[i,j]` = arrows `i→j`. Then `(A²)[i,j]` counts length-2 paths. Read off `(A²)[0,3] = 2` — the same two
paths, with no enumeration.

```python
viz.plot_matrix_pair(A, A @ A, titles=('A  (arrows i->j)', 'A^2  (length-2 paths)'))
print(f'(A^2)[0,3] = {int((A@A)[0,3])}  == path count above')
```

## Figure 3 — paths per length

How many paths `0→3` exist at each length? In the thesis this is `n_g(0,3) = dim_k(e_0·kQ_g·e_3) = (A^g)[0,3]`
— in plain words, the number of ways to get there in exactly `g` steps.

```python
lengths = [1,2,3]
counts = [int(np.linalg.matrix_power(A, g)[0,3]) for g in lengths]
viz.plot_paths_per_length(lengths, counts, 'paths 0 -> 3 by length', 'length g', 'n_g(0,3)')
print('Next (P3): each path is a message, so n_g is how many messages pile up.')
```
