# P2 — Quivers and counting paths

A **quiver** is a directed graph (multiple arrows allowed). Its superpower: we can *count paths* with simple
algebra. The key fact — **`(A^g)[i,j]` = number of length-`g` paths from `i` to `j`** — connects pictures of
graphs to the path algebra `kQ`.

**5 figures:** the diamond · each path drawn separately · `A` vs `A²` · paths-per-length · paths in a deeper graph.

```python
import numpy as np
from oversquash import viz
from aiq.quiver import Quiver
from aiq.path_algebra import PathAlgebra
Q = Quiver([0,1,2,3], [('a',0,1),('b',0,2),('c',1,3),('d',2,3)])
A = Q.adjacency_matrix()
dia = [(0,1),(0,2),(1,3),(2,3)]; pos = {0:(0,0),1:(1,1),2:(1,-1),3:(2,0)}
```

## Figure 1 — the diamond quiver

Two ways from `0` to `3` in two steps. Let's count them — first visually, then with algebra.

```python
viz.draw_graph(dia, 4, src=0, dst=3, pos=pos, title='diamond quiver')
```

## Figure 2 — count the paths, one at a time

Each length-2 path `0→3` drawn in its own colour. You can literally count them: **two**. That count is
`n_2(0,3)`.

```python
paths = [[(0,1),(1,3)], [(0,2),(2,3)]]
viz.explain_paths_highlighted(dia, pos, paths, 4, src=0, dst=3,
                              title='the two paths 0 -> 3 of length 2')
algebra_paths = PathAlgebra(Q).paths_from_to(0, 3, 2)
print('algebra agrees:', [str(p) for p in algebra_paths], '-> count =', len(algebra_paths))
```

## Figure 3 — the same count, from the matrix

`A[i,j]` = arrows `i→j`. Multiply: `(A²)[i,j]` counts length-2 paths. Read off `(A²)[0,3] = 2` — no drawing
needed. This is why matrices and path-counting are the same thing.

```python
viz.plot_matrix_pair(A, A @ A, titles=('A  (arrows i->j)', 'A^2  (length-2 paths)'))
print(f'(A^2)[0,3] = {int((A@A)[0,3])}  == the 2 paths above')
```

## Figure 4 — paths per length

How many paths `0→3` at each length? This is `n_g(0,3) = dim_k(e_0·kQ_g·e_3) = (A^g)[0,3]` — in plain words,
the number of ways to get there in exactly `g` steps.

```python
lengths = [1,2,3]
counts = [int(np.linalg.matrix_power(A, g)[0,3]) for g in lengths]
viz.plot_paths_per_length(lengths, counts, 'paths 0 -> 3 by length', 'length g', 'n_g(0,3)')
```

## Figure 5 — more structure, more paths

A 'double diamond' has **four** paths `0→6` of length 4. Drawn separately, the count is obvious — and this is
exactly the redundancy that drives over-squashing (P3).

```python
dd = [(0,1),(0,2),(1,3),(2,3),(3,4),(3,5),(4,6),(5,6)]
ddpos = {0:(0,0),1:(1,1),2:(1,-1),3:(2,0),4:(3,1),5:(3,-1),6:(4,0)}
QD = Quiver(list(range(7)), [(chr(97+i), u, v) for i,(u,v) in enumerate(dd)])
four = [[(0,1),(1,3),(3,4),(4,6)],[(0,1),(1,3),(3,5),(5,6)],
        [(0,2),(2,3),(3,4),(4,6)],[(0,2),(2,3),(3,5),(5,6)]]
viz.explain_paths_highlighted(dd, ddpos, four, 7, src=0, dst=6,
                              title='four paths 0 -> 6 of length 4')
print('Next (P3): each of these paths is a message.')
```
