# P2 — Quivers y conteo de caminos

Un **quiver** es un grafo dirigido (se permiten flechas múltiples). Su poder: podemos *contar caminos* con
álgebra limpia. El hecho clave — **`(A^g)[i,j]` = número de caminos de longitud `g` de `i` a `j`** — une los
grafos con el álgebra de caminos `kQ`.

**Figuras:** (1) el quiver diamante, (2) `A` vs `A²` como matrices, (3) gráfico de barras de caminos por
longitud.

```python
import numpy as np
from oversquash import viz
from aiq.quiver import Quiver
from aiq.path_algebra import PathAlgebra
Q = Quiver([0,1,2,3], [('a',0,1),('b',0,2),('c',1,3),('d',2,3)])
A = Q.adjacency_matrix()
```

## Figura 1 — el quiver diamante

Dos maneras de ir de `0` a `3` en dos pasos: `0→1→3` y `0→2→3`. Recuérdalo — *dos caminos*.

```python
viz.draw_graph([(0,1),(0,2),(1,3),(2,3)], 4, src=0, dst=3,
               pos={0:(0,0),1:(1,1),2:(1,-1),3:(2,0)}, title='quiver diamante')
caminos = PathAlgebra(Q).paths_from_to(0, 3, 2)
print('caminos de longitud 2 de 0 a 3:', [str(p) for p in caminos], '-> conteo =', len(caminos))
```

## Figura 2 — la matriz de adyacencia y su cuadrado

`A[i,j]` = flechas `i→j`. Entonces `(A²)[i,j]` cuenta caminos de longitud 2. Lee `(A²)[0,3] = 2` — los mismos
dos caminos, sin enumerar.

```python
viz.plot_matrix_pair(A, A @ A, titles=('A  (flechas i->j)', 'A^2  (caminos longitud 2)'))
print(f'(A^2)[0,3] = {int((A@A)[0,3])}  == conteo de caminos de arriba')
```

## Figura 3 — caminos por longitud

¿Cuántos caminos `0→3` hay en cada longitud? En la tesis esto es `n_g(0,3) = dim_k(e_0·kQ_g·e_3) = (A^g)[0,3]`
— en palabras simples, el número de formas de llegar en exactamente `g` pasos.

```python
longitudes = [1,2,3]
conteos = [int(np.linalg.matrix_power(A, g)[0,3]) for g in longitudes]
viz.plot_paths_per_length(longitudes, conteos, 'caminos 0 -> 3 por longitud', 'longitud g', 'n_g(0,3)')
print('Siguiente (P3): cada camino es un mensaje, así que n_g es cuántos mensajes se acumulan.')
```
