# P2 — Quivers y conteo de caminos

Un **quiver** es un grafo dirigido (se permiten flechas múltiples). Su superpoder: podemos *contar caminos* con
álgebra simple. El hecho clave — **`(A^g)[i,j]` = número de caminos de longitud `g` de `i` a `j`** — conecta
las imágenes de grafos con el álgebra de caminos `kQ`.

**5 figuras:** el diamante · cada camino por separado · `A` vs `A²` · caminos por longitud · caminos en un
grafo más profundo.

```python
import numpy as np
from oversquash import viz
from aiq.quiver import Quiver
from aiq.path_algebra import PathAlgebra
Q = Quiver([0,1,2,3], [('a',0,1),('b',0,2),('c',1,3),('d',2,3)])
A = Q.adjacency_matrix()
dia = [(0,1),(0,2),(1,3),(2,3)]; pos = {0:(0,0),1:(1,1),2:(1,-1),3:(2,0)}
```

## Figura 1 — el quiver diamante

Dos maneras de ir de `0` a `3` en dos pasos. Contémoslas — primero visualmente, luego con álgebra.

```python
viz.draw_graph(dia, 4, src=0, dst=3, pos=pos, title='quiver diamante')
```

## Figura 2 — cuenta los caminos, uno por uno

Cada camino `0→3` de longitud 2 dibujado en su propio color. Puedes contarlos literalmente: **dos**. Ese
conteo es `n_2(0,3)`.

```python
caminos = [[(0,1),(1,3)], [(0,2),(2,3)]]
viz.explain_paths_highlighted(dia, pos, caminos, 4, src=0, dst=3,
                              title='los dos caminos 0 -> 3 de longitud 2', path_fmt='camino {i}')
caminos_alg = PathAlgebra(Q).paths_from_to(0, 3, 2)
print('el álgebra coincide:', [str(p) for p in caminos_alg], '-> conteo =', len(caminos_alg))
```

## Figura 3 — el mismo conteo, desde la matriz

`A[i,j]` = flechas `i→j`. Multiplica: `(A²)[i,j]` cuenta caminos de longitud 2. Lee `(A²)[0,3] = 2` — sin
dibujar. Por esto las matrices y el conteo de caminos son lo mismo.

```python
viz.plot_matrix_pair(A, A @ A, titles=('A  (flechas i->j)', 'A^2  (caminos longitud 2)'))
print(f'(A^2)[0,3] = {int((A@A)[0,3])}  == los 2 caminos de arriba')
```

## Figura 4 — caminos por longitud

¿Cuántos caminos `0→3` hay en cada longitud? Esto es `n_g(0,3) = dim_k(e_0·kQ_g·e_3) = (A^g)[0,3]` — en
palabras simples, el número de formas de llegar en exactamente `g` pasos.

```python
longitudes = [1,2,3]
conteos = [int(np.linalg.matrix_power(A, g)[0,3]) for g in longitudes]
viz.plot_paths_per_length(longitudes, conteos, 'caminos 0 -> 3 por longitud', 'longitud g', 'n_g(0,3)')
```

## Figura 5 — más estructura, más caminos

Un 'doble diamante' tiene **cuatro** caminos `0→6` de longitud 4. Dibujados por separado, el conteo es obvio —
y esta es justo la redundancia que provoca el over-squashing (P3).

```python
dd = [(0,1),(0,2),(1,3),(2,3),(3,4),(3,5),(4,6),(5,6)]
ddpos = {0:(0,0),1:(1,1),2:(1,-1),3:(2,0),4:(3,1),5:(3,-1),6:(4,0)}
cuatro = [[(0,1),(1,3),(3,4),(4,6)],[(0,1),(1,3),(3,5),(5,6)],
          [(0,2),(2,3),(3,4),(4,6)],[(0,2),(2,3),(3,5),(5,6)]]
viz.explain_paths_highlighted(dd, ddpos, cuatro, 7, src=0, dst=6,
                              title='cuatro caminos 0 -> 6 de longitud 4', path_fmt='camino {i}')
print('Siguiente (P3): cada uno de estos caminos es un mensaje.')
```
