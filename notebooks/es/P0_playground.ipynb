# P0 — Patio de juegos: grafos pequeños

Un calentamiento práctico. Dibujamos grafos pequeños y contamos sus **caminos**. Todo el proyecto descansa en
un hecho: *cuántos caminos conectan dos nodos* decide cuánta información se aprieta junta. Edita las listas de
aristas y vuelve a ejecutar — todo se actualiza.

**Verás 3+ figuras:** un grafo línea, un grafo de dos rutas y un diamante (más un doble diamante).

```python
import numpy as np, torch
from oversquash import viz
from oversquash.ideal_bridge import walk_operators

def conteo_caminos(edges, n, src, dst, max_len=4):
    raw, eff = walk_operators(np.array(edges).T, n, max_length=max_len)
    for g in range(max_len):
        r, e = int(raw[g][src, dst]), int(eff[g][src, dst])
        if r:
            print(f'  longitud {g+1}: {r} camino(s)' + ('' if r == e else f'  (kQ/I de-duplica -> {e})'))
```

## Figura 1 — una línea recta (sin redundancia)

`0→1→2→3→4`: exactamente **un** camino de punta a punta. Lejos, pero sin acumulación — así que una GNN normal
puede (lentamente) llevar la señal. El over-squashing necesita *muchos* caminos en competencia, que una línea
no tiene.

```python
linea = [(0,1),(1,2),(2,3),(3,4)]
viz.draw_graph(linea, 5, src=0, dst=4, pos={i:(i,0) for i in range(5)}, title='línea: un camino 0->4')
conteo_caminos(linea, 5, 0, 4)
```

## Figura 2 — dos rutas de distinta longitud

Una arista corta `0→3` y un desvío largo `0→1→2→3`. La información llega a `3` en **dos rangos distintos**
(longitud 1 y longitud 3). Por esto indexamos todo por el número de saltos `g`.

```python
dos = [(0,3),(0,1),(1,2),(2,3)]
viz.draw_graph(dos, 4, src=0, dst=3, pos={0:(0,0),1:(1,1),2:(2,1),3:(3,0)}, title='dos rutas 0->3')
conteo_caminos(dos, 4, 0, 3)
```

## Figura 3 — rutas paralelas, misma longitud (redundancia)

El 'diamante': `0→1→3` y `0→2→3`, ambos de longitud 2. Dos caminos, pero si los nodos del medio llevan el
*mismo* contenido son **redundantes** — mira cómo `kQ/I` colapsa el conteo de 2 a 1.

```python
diamante = [(0,1),(0,2),(1,3),(2,3)]
viz.draw_graph(diamante, 4, src=0, dst=3, pos={0:(0,0),1:(1,1),2:(1,-1),3:(2,0)}, title='diamante: 2 caminos paralelos')
conteo_caminos(diamante, 4, 0, 3)
```

## Figura 4 — tu turno: un doble diamante

Dos diamantes en fila dan **4** caminos paralelos de longitud 4. Edita `mi_grafo` y vuelve a ejecutar.

```python
mi_grafo = [(0,1),(0,2),(1,3),(2,3),(3,4),(3,5),(4,6),(5,6)]
viz.draw_graph(mi_grafo, 7, src=0, dst=6, title='doble diamante: 4 caminos paralelos 0->6')
conteo_caminos(mi_grafo, 7, 0, 6, max_len=4)
print('\nMás caminos paralelos de la misma longitud = peor over-squashing. ¡Sigamos con P1!')
```
